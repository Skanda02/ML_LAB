```python
# Import the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data = pd.DataFrame({
    'Age': [23,23,27,27,39,41,47,50,52,54,54,56,67,68,58,60,61],
    '%Fat': [7.8,9.5,17.8,18.5,19.8,20.5,21.8,22.5,23.8,24.5,25.8,26.5,27.8,28.5,29.8,30.5,31.8]
})

# Save it to a CSV file
data.to_csv('age_fat_data.csv', index=False)
```

```python
# Lode the data from the csv file
data = pd.read_csv('age_fat_data.csv')
```

```python
data.head()
```

```python
nummerric_column = "%Fat"
data_num = data[nummerric_column]
data_num.head()
```

```python
# Mean
mean_value = data_num.mean()
print("Mean:", mean_value)
```

```python
# median
median_value = data_num.median()
print("Median:", median_value)
```

```python
# Mode
Mode_val = data_num.mode()
print("Mode:", Mode_val)
```

```python
# Mode
Mode_val = data_num.mode()[0]
print("Mode:", Mode_val)
```

```python
# Std_deviation

std_dev = data_num.std()
print("Std_deviation:", std_dev)
```

```python
# Variance 
variance = data_num.var()
print("Variance:", variance)
```

```python
# Range value
rang_val = data_num.max() - data_num.min()
print("Range:",rang_val)
```

## Histogram

```python
plt.figure(figsize=(8,5))
plt.hist(data_num,bins=10,color="blue",edgecolor="black")
plt.title("Histogram of % Fat")
plt.xlabel("%Fat")
plt.ylabel("Frequence")
plt.show()
```

## Box plot :

```python
plt.figure(figsize=(8,5))
sns.boxplot(x= data_num, color="skyblue")
plt.title("Box plot of %Fat")
plt.show()
```

## Identify outliers usin IQR

```python
q1 = data_num.quantile(0.25)
q3 = data_num.quantile(0.75)

iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

outliers = data_num[(data_num < lower_bound) | (data_num > upper_bound)]
print("\n Outliers:")
print(outliers)
```

```python
def age_group(age):
    if age < 30:
        return "Young"
    elif age <= age <= 50:
        return "Middle_aged"
    else:
        return "older"
```

```python
data['Age_group'] = data["Age"].apply(age_group)
```

```python
data.head()
```

```python
categorical_column = "Age_group"
data_cat = data[categorical_column]
```

```python
## Compute feequency of each category
```

```python
category_counts = data_cat.value_counts()
category_counts
```

# Bar Chart for the categorical column

```python
plt.figure(figsize=(8,5))
category_counts.plot(kind='bar',color='blue',edgecolor='black')
plt.xlabel("Age_group")
plt.ylabel("Frequency")
plt.show()
```

# Pie Chart for Categorical column

```python
plt.figure(figsize=(8,5))
category_counts.plot(kind='pie',autopct='%1.1f%%',startangle=90,colors=sns.color_palette('pastel'))
plt.title("Pie chart for age_groups")
plt.ylabel(" ")
plt.show()
```
